```python
import joblib
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from pathlib import Path


pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

%matplotlib inline
%config IPCompleter.use_jedi = False
```

```python
manzanas_path = Path('../data/external/shapefiles/').resolve()
```

```python
def cargar_shape(path: str):
    df = gpd.read_file(path)
    df = df.drop('TIPOMZA', axis=1)
    cvegeo = df.CVE_ENT + df.CVE_MUN + df.CVE_LOC + df.CVE_AGEB + df.CVE_MZA
    df = df.assign(cvegeo=cvegeo)
    df = df.drop(['CVE_ENT', 'CVE_MUN', 'CVE_LOC', 'CVE_AGEB', 'CVE_MZA'], axis=1)
    df = df.to_crs({'init': 'epsg:4326'})
    return df

dfs = joblib.Parallel(n_jobs=-1, verbose=11)(
    joblib.delayed(cargar_shape)(path.as_posix())
    for path in manzanas_path.glob('*/conjunto de datos/[0-9][0-9]m.shp')
)

df_manzanas = pd.concat(dfs, axis=0, ignore_index=True)

print(df_manzanas.shape)
df_manzanas.head()
```

```python
df_manzanas.to_file('../data/external/manzanas_shapefile/')
```

```python
from shapely.geometry import Point
```

```python
# Se cargan las subagencias
df_catalogo = pd.read_parquet('../data/processed/Catalogo_Subagencias.parquet', engine='pyarrow')
geometry = [
    Point(row.Longitud, row.Latitud)
    for row in df_catalogo.itertuples()]
df_catalogo = df_catalogo.assign(geometry=geometry)
df_catalogo = gpd.GeoDataFrame(df_catalogo, crs={'init': 'epsg:4326'})
df_catalogo.head()
```

```python
df_join = gpd.sjoin(df_catalogo, df_manzanas, how='inner', op='intersects')
```
