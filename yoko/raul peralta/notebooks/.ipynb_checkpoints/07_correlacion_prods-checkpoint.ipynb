```python

import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

%matplotlib inline
%config IPCompleter.use_jedi = False
```

```python
def load_historical(path: str):
    df_hist = pd.read_parquet(path, engine='pyarrow')
    df_total_skus = df_hist.groupby('SKU', as_index=False).Hectolitros.sum().sort_values('Hectolitros')
    skus_cero = df_total_skus[df_total_skus.Hectolitros == 0].SKU.unique()
    df_hist_filtrado = df_hist.loc[~df_hist.SKU.isin(skus_cero)]
    # print(df_hist_filtrado.shape)
    return df_hist_filtrado
    
historical_path = '../data/processed/HistoricoVentas.parquet'
df_hist = load_historical(historical_path)
print(df_hist.shape)
df_hist.head()
```

```python
df_total_sku = df_hist.groupby('SKU', as_index=False).Hectolitros.sum().sort_values('Hectolitros', ascending=False)
df_total_sku = df_total_sku.assign(pc_total=(df_total_sku.Hectolitros.divide(df_total_sku.Hectolitros.sum())* 100))
df_total_sku = df_total_sku.reset_index(drop=True)
skus_grandes = df_total_sku.head(17).SKU.unique()
```

```python
# df_total_sku.pc_total.cumsum()
```

```python
subagencias = df_hist.Subagencia.unique()

corrs = []
for sub_ag in subagencias:
    df_aux = df_hist.loc[df_hist.Subagencia == sub_ag]
    df_aux = df_aux.loc[df_aux.SKU.isin(skus_grandes)]
    df_aux = df_aux.pivot(index='Mes', columns='SKU', values='Hectolitros')
    df_corr = df_aux.corr()
    cond = np.triu(np.ones(df_corr.shape)).astype(np.bool)
    df_corr = df_corr.where(cond)
    df_corr.columns.name = 'SKU_1'
    df_corr.index.name = 'SKU_2'
    df_corr = df_corr.stack().reset_index()
    df_corr = df_corr.rename(columns={0: 'correlacion'})
    indice = df_corr.SKU_1.str.cat(df_corr.SKU_2, sep='-')
    df_corr = df_corr.assign(indice=indice).set_index('indice').drop(['SKU_1', 'SKU_2'], axis=1)
    corrs.append(df_corr)
```

```python
df_corrs = pd.concat(corrs, axis=1)
df_corrs.head()
```

```python
df_final = df_corrs.mean(axis=1).sort_values(ascending=False)
df_final = df_final.reset_index()
SKU_1 = df_final['index'].map(lambda x: x.split('-')[0])
SKU_2 = df_final['index'].map(lambda x: x.split('-')[1])
df_final = df_final.assign(SKU_1=SKU_1, SKU_2=SKU_2)
```

```python
import matplotlib.pyplot as plt
with plt.style.context('seaborn-talk'):
    import seaborn as sns
    fig, ax = plt.subplots(figsize=(20, 15))
    df_heat = df_final.pivot(index='SKU_1', columns='SKU_2', values=0)
    sns.heatmap(df_heat, cmap='viridis', annot=True)
    plt.tight_layout()
```

```python
def cargar_denue(path):
    cols = [
        'id',
         'nom_estab',
         'raz_social',
         'codigo_act',
         'nombre_act',
         'per_ocu',
         'tipo_vial',
         'nom_vial',
         'tipo_v_e_1',
         'nom_v_e_1',
         'tipo_v_e_2',
         'nom_v_e_2',
         'tipo_v_e_3',
         'nom_v_e_3',
         'numero_ext',
         'letra_ext',
         'edificio',
         'edificio_e',
         'numero_int',
         'letra_int',
         'tipo_asent',
         'nomb_asent',
         'tipoCenCom',
         'nom_CenCom',
         'num_local',
         'cve_ent',
         'cve_mun',
         'cve_loc',
         'ageb',
         'manzana',
         'tipoUniEco',
         'latitud',
         'longitud',
         'fecha_alta'
    ]
    df = pd.read_csv(
        path, error_bad_lines=False, usecols=cols,
        dtype={'codigo_act': str, 'cve_ent': str, 'cve_mun': str, 'cve_loc': str, 'ageb': str, 'manzana': str}
    )
    cvegeo = df.cve_ent + df.cve_mun + df.cve_loc + df.ageb + df.manzana
    df = df.assign(cvegeo=cvegeo)
    df = df.drop(['cve_ent', 'cve_mun', 'cve_loc', 'ageb', 'manzana'], axis=1)
    return df

denue_path = Path('../data/external/denue/').resolve()


dfs = joblib.Parallel(n_jobs=-1, verbose=0)(
    joblib.delayed(cargar_denue)(path)
    for path in denue_path.glob('*/conjunto_de_datos/*.csv')
)

df_denue = pd.concat(dfs, axis=0, ignore_index=True)
df_denue = df_denue.loc[~df_denue.cvegeo.isnull()]
print(df_denue.shape)
df_denue.head()
```

```python
df_denue.to_csv('../data/external/denue_completo.csv', encoding='utf-8', index=False, quoting=1)
```
