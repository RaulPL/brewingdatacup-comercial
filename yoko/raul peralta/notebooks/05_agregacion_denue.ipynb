```python
import joblib
import pyproj
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import geopandas as gpd
from shapely.geometry import Point

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

%matplotlib inline
%config IPCompleter.use_jedi = False
```

```python
# Se carga denue

denue_cols = [
    'id',
    'nom_estab',
    'raz_social',
    'codigo_act',
    'nombre_act',
    'per_ocu',
    'nomb_asent',
    'tipoCenCom',
    'nom_CenCom',
    'num_local',
    'tipoUniEco',
    'latitud',
    'longitud',
    'fecha_alta',
    'cvegeo'
]

df_denue = pd.read_csv('../data/external/denue_completo.csv', dtype={'cvegeo': str, 'codigo_act': str}, usecols=denue_cols)
print(df_denue.shape)


scian_cats = {
    '461212': 'Comercio al por menor de cerveza',
    '431213': 'Comercio al por mayor de cerveza',
    '431110': 'Comercio al por mayor de abarrotes',
    '7224': 'Centros nocturnos, bares, cantinas y similares',
    '722511': 'Restaurantes con servicio de preparación de alimentos a la carta o de comida corrida',
    '462112': 'Comercio al por menor en minisupers (Oxxo) ',
    '462111': 'Comercio al por menor en supermercados',
    
}

dfs = {k: df_denue.loc[df_denue.codigo_act.str.contains(k)].assign(categoria=k) for k in scian_cats}
df_denue_filt = pd.concat(dfs.values(), axis=0, ignore_index=True)
print(df_denue_filt.shape)
del df_denue
df_denue_filt.head()
```

```python
df_catalogo = pd.read_parquet('../data/processed/Catalogo_Subagencias.parquet', engine='pyarrow')
# se filtra subagencia en el mar
df_catalogo = df_catalogo.loc[df_catalogo.Subagencia != '10209']
df_catalogo = df_catalogo.loc[df_catalogo.Subagencia != '10059']

geometry = [
    Point(row.Longitud, row.Latitud)
    for row in df_catalogo.itertuples()]
df_catalogo = df_catalogo.assign(geometry=geometry)
df_catalogo = gpd.GeoDataFrame(df_catalogo, crs={'init': 'epsg:4326'})

# se genera
RADIO = 5000
circulo = (df_catalogo.geometry.to_crs({'init': 'epsg:3857'})
                      .buffer(RADIO)
                      .to_crs({'init': 'epsg:4326'}))

df_catalogo = df_catalogo.assign(geometry=circulo)

df_catalogo.head()
```

# Denue GeoDataFrame

```python
puntos_denue = [
    Point(row.longitud, row.latitud)
    for row in df_denue_filt.itertuples()
]

df_denue_filt = df_denue_filt.assign(geometry=puntos_denue)

df_denue_filt = gpd.GeoDataFrame(df_denue_filt, crs={'init': 'epsg:4326'})
df_denue_filt.head()
```

```python
df_join = gpd.sjoin(df_catalogo[['Subagencia', 'geometry']], df_denue_filt, how='inner', op='intersects')
print(df_join.shape)
df_join.head()
```

```python
df_conteo = df_join.groupby(['Subagencia', 'categoria'], as_index=False).cvegeo.count()
df_conteo = df_conteo.rename(columns={'cvegeo': 'num_lugares'})
print(df_conteo.shape)
df_conteo.head()
```

```python
df_categorias = df_conteo.pivot(index='Subagencia', columns='categoria', values='num_lugares').fillna(0).reset_index()
print(df_categorias.shape)
df_categorias.head()
```

```python
df_categorias = df_categorias.rename(
    columns={'461212': 'comercio_al_por_menor_cerveza', 
             '462112': 'minisupers', 
             '7224': 'bares_cantinas', 
             '431110': 'comercio_al_por_mayor_abarrotes'}
)
# df_categorias
```

```python
def load_historical(path: str):
    df_hist = pd.read_parquet(path, engine='pyarrow')
    df_total_skus = df_hist.groupby('SKU', as_index=False).Hectolitros.sum().sort_values('Hectolitros')
    skus_cero = df_total_skus[df_total_skus.Hectolitros == 0].SKU.unique()
    df_hist_filtrado = df_hist.loc[~df_hist.SKU.isin(skus_cero)]
    # print(df_hist_filtrado.shape)
    return df_hist_filtrado
    
historical_path = '../data/processed/HistoricoVentas.parquet'
df_hist = load_historical(historical_path)
print(df_hist.shape)

df_dependiente = df_hist.groupby(['Subagencia', 'SKU'], as_index=False).Hectolitros.mean().sort_values('Hectolitros', ascending=False)
df_total = df_dependiente.groupby('Subagencia', as_index=False).Hectolitros.sum()
df_dependiente = pd.merge(df_total, df_catalogo, on='Subagencia', how='inner')
```
